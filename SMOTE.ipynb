## SMOTE(Synthetic Minority oversampling technique )

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],random_state=1)
```

```python
#x is independent feature
#y dependent feature
x
```

```python
y
```

```python
x,y
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=["target"])
df3=pd.concat([df1,df2],axis=1)
```

```python
df3
```

```python
df3['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df3['f1'],df3['f2'],c=df3['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
#transform the dataset
oversample=SMOTE()
x,y=oversample.fit_resample(df3[['f1','f2']],df3['target'])
```
